```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import itertools
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors

# 1. Load the datasets
walmart_products = df # pd.read_csv("/content/walmart_data/walmart_data.csv")  # Adjust path if needed
walmart_sales = pd.read_csv("/root/.cache/kagglehub/datasets/devarajv88/walmart-sales-dataset/versions/1/walmart.csv")  # Adjust path if needed

# 2. Preprocessing walmart_products
# Handle missing values:
# Fill missing 'Brand' with 'Unknown' or the most frequent brand
walmart_products['Brand'].fillna('Unknown', inplace=True)
# Fill missing 'Category' with the most frequent category
walmart_products['Category'].fillna(walmart_products['Category'].mode()[0], inplace=True)

# Drop 'Package Size' and 'Postal Code' because they have all missing values
# walmart_products.drop(['Package Size', 'Postal Code'], axis=1, inplace=True)

#For the 'Item Number' column, it is better to replace NaN values with the mean or median because dropping it can remove relevant data and since it is related to products.
# Calculate the mean of 'Item Number'
item_number_mean = walmart_products['Item Number'].mean()

# Fill NaN values in 'Item Number' with the mean
walmart_products['Item Number'].fillna(item_number_mean, inplace=True)


# 3. Preprocessing walmart_sales
# Rename columns for consistency (optional)
walmart_sales.rename(columns={'User_ID': 'UserID', 'Product_ID': 'ProductID', 'Product_Category': 'ProductCategory'}, inplace=True, errors = 'ignore')

# Handle categorical variables (if needed) -  e.g., City_Category if you plan to use it.
# Example: One-hot encode City_Category
#walmart_sales = pd.get_dummies(walmart_sales, columns=['City_Category'], drop_first=True)

# Print info to verify
print("Walmart Products Info:")
walmart_products.info()
print("\nWalmart Sales Info:")
walmart_sales.info()

# 4. Combine product name and description for embedding
walmart_products['combined_text'] = walmart_products['Product Name'] + " " + walmart_products['Description']
walmart_products['combined_text'].fillna('', inplace=True)  # Handle any remaining NaNs

# 5. Load a pre-trained Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2') # A good balance of speed and accuracy

# 6. Generate embeddings for all products
product_embeddings = model.encode(walmart_products['combined_text'].tolist())

print("Product Embeddings Shape:", product_embeddings.shape)

# 7. Create a user-item interaction matrix
user_item_matrix = walmart_sales.pivot_table(index='UserID', columns='ProductID', values='Purchase', fill_value=0)

# 8. Split data into training and testing sets
train_data, test_data = train_test_split(user_item_matrix, test_size=0.2, random_state=42)

# 9. Train a KNN model
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(train_data)

def get_user_recommendations(user_id, num_recommendations=5):
    """
    Recommends products to a user based on collaborative filtering.
    """
    try:
      # Find the user's vector
      user_vector = user_item_matrix.loc[user_id].values.reshape(1, -1)

      # Find similar users
      distances, indices = knn_model.kneighbors(user_vector, n_neighbors=11)  # Increased n_neighbors

      # Aggregate the purchase history of similar users
      similar_users = user_item_matrix.iloc[indices[0]].mean()

      # Exclude products the user has already purchased
      user_purchases = user_item_matrix.loc[user_id]
      similar_users[user_purchases > 0] = 0

      # Get top recommendations
      recommendations = similar_users.sort_values(ascending=False).head(num_recommendations)

      return recommendations.index.tolist()
    except KeyError:
      return "User not found"

def get_hybrid_recommendations(user_id, current_cart_items, num_recommendations=5, alpha=0.5):
    """
    Combines collaborative filtering and content-based (product similarity) recommendations.
    """

    # 1. Collaborative Filtering Recommendations
    cf_recommendations = get_user_recommendations(user_id, num_recommendations=num_recommendations)

    # 2. Content-Based (Product Similarity) Recommendations
    cb_recommendations = []
    if current_cart_items:
        # Embed the current cart items, handling potential errors
        cart_embeddings = []
        for item in current_cart_items:
            matching_products = walmart_products[walmart_products['Product Name'] == item]
            if not matching_products.empty:
                cart_embeddings.append(matching_products['combined_text'].iloc[0])

        # Proceed with content-based recommendations if cart_embeddings is not empty
        if cart_embeddings:
            cart_embeddings = model.encode(cart_embeddings)

            # Calculate similarity to all products
            similarity_scores = cosine_similarity(cart_embeddings, product_embeddings).mean(axis=0)  # Average similarity across cart items

            # Get top similar products, excluding those already in the cart
            similar_product_indices = np.argsort(similarity_scores)[::-1]
            for idx in similar_product_indices:
                product_name = walmart_products['Product Name'].iloc[idx]
                if product_name not in current_cart_items:
                    cb_recommendations.append(product_name)
                    if len(cb_recommendations) >= num_recommendations:
                        break

    # 3. Combine Recommendations (Weighted Average)
    hybrid_recommendations = []

    if cf_recommendations != "User not found":
        hybrid_recommendations.extend(cf_recommendations[:int(num_recommendations * alpha)])

    hybrid_recommendations.extend(cb_recommendations[:int(num_recommendations * (1 - alpha))])

    return hybrid_recommendations[:num_recommendations]  #Ensure no more than the required recommendations.

def get_shop_priority_items(priority=""):
    """
    Placeholder function to simulate getting shop owner-prioritized items.
    In a real application, this would access a database or API.
    """
    if priority == "expiring_soon":
        # Simulated logic:  Return a few items with low item numbers, representing old inventory.
        shop_priority_items = walmart_products[walmart_products['Item Number'] < walmart_products['Item Number'].quantile(0.2)]['Product Name'].tolist()[:3]  #Top 3
        #This has the code to show what to do if priority includes expiring soon, high levels, and promotion and sale
    elif priority == "high_stock_levels":
        shop_priority_items = walmart_products[walmart_products['Item Number'] > walmart_products['Item Number'].quantile(0.8)]['Product Name'].tolist()[:3] #Top 3

    elif priority == "promotion_sale":
          shop_priority_items = walmart_products[(walmart_products['List Price'] > walmart_products['Sale Price'])]['Product Name'].tolist()[:3]
    else:
        shop_priority_items = [] # No priority specified

    return shop_priority_items

def find_best_gap_filling_combination(gap_filling_pool, gap, walmart_products):
    """
    Finds the best combination of items from the gap_filling_pool to reach the gap amount.
    Prioritizes exact matches and then minimizes the difference.
    """
    best_combination = None
    min_difference = float('inf')

    # 1. Check for exact match
    for item in gap_filling_pool:
        product_data = walmart_products[walmart_products['Product Name'] == item]['Sale Price']
        if not product_data.empty:
            item_price = product_data.iloc[0]
            if abs(item_price - gap) < 1e-6:
                return [item]  # Found an exact match!

    # 2. Check combinations of items with size (1-3). If you add more items, it can increase computation
    for r in range(1, min(4, len(gap_filling_pool) + 1)):
        for combination in itertools.combinations(gap_filling_pool, r):
            total_price = 0
            valid_combination = True
            for item in combination:
                product_data = walmart_products[walmart_products['Product Name'] == item]['Sale Price']

                if not product_data.empty:
                    item_price = product_data.iloc[0]
                    total_price += item_price #Safe add here
                else:
                    valid_combination = False
                    break

            if not valid_combination:
                continue

            difference = abs(total_price - gap)

            if difference < min_difference:
                min_difference = difference
                best_combination = list(combination)

    return best_combination

def get_smart_cart_recommendations(user_id, current_cart_items, shipping_threshold=20, alpha=0.5, shop_priority="", max_overspend_percentage=0.2):
    """
    Combines collaborative filtering, content-based recommendations, and shop priority items
    while considering the shipping threshold and a maximum overspend percentage.

    shop_priority: A string indicating the shop owner's priority (e.g., "expiring_soon", "high_stock_levels", "promotion_sale" , "").
    max_overspend_percentage: The maximum percentage the user is allowed to overspend the initial 20$ (shipping threshold amount).

    """

    # 1. Hybrid Recommendations (CF + Content-Based)
    hybrid_recommendations = get_hybrid_recommendations(user_id, current_cart_items, num_recommendations=5, alpha=alpha)

    # 2. Calculate current cart value

    current_cart_value = 0
    #Safe iteration here
    for item in current_cart_items:
      product_data = walmart_products[walmart_products['Product Name'] == item]['Sale Price'] # Gets the product name from there

      if not product_data.empty:
          current_cart_value += product_data.iloc[0] # adds price


    # 3. Calculate gap to reach shipping threshold
    gap = shipping_threshold - current_cart_value

    # 4. Calculate max acceptable cart value
    max_cart_value = shipping_threshold * (1 + max_overspend_percentage) # for our example, is 24. If you want over the max, you can recommend it as well! This needs to be user facing

    # 5. Get shop priority items
    shop_priority_items = get_shop_priority_items(shop_priority)

    # 6. Create Gap-Filling Pool
    gap_filling_pool = walmart_products[walmart_products['Sale Price'] <= gap]['Product Name'].tolist()

    # 7. Final Recommendations List
    final_recommendations = []

    # 8. Attempt to Fill the Gap with Exact or Combinations
    best_gap_fill_items = find_best_gap_filling_combination(gap_filling_pool, gap, walmart_products)

    if best_gap_fill_items:
        final_recommendations.extend(best_gap_fill_items)

    # 9. Add Shop Priority Items (if any and within budget)

    for item in shop_priority_items:
        product_data = walmart_products[walmart_products['Product Name'] == item]['Sale Price'] # To check the product

        if not product_data.empty: # Not empty

            item_price = product_data.iloc[0]
            total_recommedations = 0

            for i in final_recommendations:
              rec = walmart_products[walmart_products['Product Name'] == i]['Sale Price'] #Again. Check value if exists

              if not rec.empty:
                  total_recommedations = rec.iloc[0]  #Safe, always has the value
            #Added if it is bigger max

            if (current_cart_value + total_recommedations + item_price) <= max_cart_value:

                final_recommendations.append(item) # If all prices does not exceed add.
            else:
                # The goal to get to the value of the max_cart_value is to not happen.
                # This is to recommend.

                print("You are about to exceed the shipping threshold (20$) by too much!") # if not, print

    # 10. If the amount is not enough to meet, this will add recommend from hybrid. It is a TODO to implement. #TODO - If not good

    return final_recommendations

#Example Usage
user_id = walmart_sales['UserID'].iloc[0] # Example user
current_cart = ['Duracell - AAA Batteries (4-Pack)'] # Example cart
shipping_threshold = 20  # Example shipping threshold
shop_priority = "promotion_sale"  # Example shop priority
max_overspend_percentage = 0.2 # Example overspend allowed. Can have a slider

recommendations = get_smart_cart_recommendations(user_id, current_cart, shipping_threshold, shop_priority = shop_priority, max_overspend_percentage = max_overspend_percentage) # This the function in todo implement (10. to add items from hybrid recommendations)
print(f"Smart Cart Recommendations for user {user_id} with cart {current_cart}: {recommendations}") # To recommend the products for the example
```

```python




import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import itertools
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors

# 1. Load the datasets
walmart_products = df  # pd.read_csv("/content/walmart_data/walmart_data.csv")  # Adjust path if needed
walmart_sales = pd.read_csv("/root/.cache/kagglehub/datasets/devarajv88/walmart-sales-dataset/versions/1/walmart.csv")  # Adjust path if needed

# 2. Preprocessing walmart_products
# Handle missing values:
# Fill missing 'Brand' with 'Unknown' or the most frequent brand
walmart_products['Brand'].fillna('Unknown', inplace=True)
# Fill missing 'Category' with the most frequent category
walmart_products['Category'].fillna(walmart_products['Category'].mode()[0], inplace=True)

# Drop 'Package Size' and 'Postal Code' because they have all missing values
# walmart_products.drop(['Package Size', 'Postal Code'], axis=1, inplace=True)

# For the 'Item Number' column, it is better to replace NaN values with the mean or median
item_number_mean = walmart_products['Item Number'].mean()
walmart_products['Item Number'].fillna(item_number_mean, inplace=True)

# 3. Preprocessing walmart_sales
print(walmart_sales.columns)  # For reference
walmart_sales.rename(columns={
    'User_ID': 'UserID',
    'Product_ID': 'ProductID',
    'Product_Category': 'ProductCategory'
}, inplace=True, errors='ignore')

# 4. Combine product name and description for embedding
walmart_products['combined_text'] = walmart_products['Product Name'] + " " + walmart_products['Description']
walmart_products['combined_text'].fillna('', inplace=True)  # Handle any remaining NaNs

# 5. Load a pre-trained Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')  # A good balance of speed and accuracy

# 6. Generate embeddings for all products
product_embeddings = model.encode(walmart_products['combined_text'].tolist())
print("Product Embeddings Shape:", product_embeddings.shape)

# 7. Create a user-item interaction matrix
user_item_matrix = walmart_sales.pivot_table(
    index='UserID', columns='ProductID',
    values='Purchase', fill_value=0
)

# 8. Split data into training and testing sets
train_data, test_data = train_test_split(user_item_matrix, test_size=0.2, random_state=42)

# 9. Train a KNN model
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(train_data)

def get_user_recommendations(user_id, num_recommendations=5):
    """
    Recommends products to a user based on collaborative filtering.
    """
    try:
        # Find the user's vector
        user_vector = user_item_matrix.loc[user_id].values.reshape(1, -1)

        # Find similar users
        distances, indices = knn_model.kneighbors(user_vector, n_neighbors=11)  # Increased n_neighbors

        # Aggregate the purchase history of similar users
        similar_users = user_item_matrix.iloc[indices[0]].mean()

        # Exclude products the user has already purchased
        user_purchases = user_item_matrix.loc[user_id]
        similar_users[user_purchases > 0] = 0

        # Get top recommendations
        recommendations = similar_users.sort_values(ascending=False).head(num_recommendations)

        return recommendations.index.tolist()
    except KeyError:
        return "User not found"


def get_hybrid_recommendations(user_id, current_cart_items, num_recommendations=5, alpha=0.5):
    """
    Combines collaborative filtering and content-based (product similarity) recommendations.
    """
    # 1. Collaborative Filtering Recommendations
    cf_recommendations = get_user_recommendations(user_id, num_recommendations=num_recommendations)

    # 2. Content-Based (Product Similarity) Recommendations
    cb_recommendations = []
    if current_cart_items:
        # Embed the current cart items
        cart_embeddings = []
        for item in current_cart_items:
            matching_products = walmart_products[walmart_products['Product Name'] == item]
            if not matching_products.empty:
                cart_embeddings.append(matching_products['combined_text'].iloc[0])
            else:
                print(f"Warning: Product '{item}' not found in walmart_products. Skipping embedding.")

        # Proceed with content-based recommendations if cart_embeddings is not empty
        if cart_embeddings:
            cart_embeddings = model.encode(cart_embeddings)
            # Calculate similarity to all products (average similarity across cart items)
            similarity_scores = cosine_similarity(cart_embeddings, product_embeddings).mean(axis=0)

            # Get top similar products, excluding those already in the cart
            similar_product_indices = np.argsort(similarity_scores)[::-1]
            for idx in similar_product_indices:
                product_name = walmart_products['Product Name'].iloc[idx]
                if product_name not in current_cart_items:
                    cb_recommendations.append(product_name)
                    if len(cb_recommendations) >= num_recommendations:
                        break

    # 3. Combine Recommendations (Weighted Average)
    hybrid_recommendations = []
    if cf_recommendations != "User not found":
        hybrid_recommendations.extend(cf_recommendations[:int(num_recommendations * alpha)])
    else:
        print("There is no user for collaborative filtering.")

    hybrid_recommendations.extend(cb_recommendations[:int(num_recommendations * (1 - alpha))])

    return hybrid_recommendations[:num_recommendations]


def get_shop_priority_items(priority=""):
    """
    Dynamically retrieves shop-prioritized items based on stock levels, expiration, or promotions.
    """
    if priority == "expiring_soon":
        # Select items with the lowest "Item Number" (oldest inventory)
        shop_priority_items = walmart_products[
            walmart_products['Item Number'] < walmart_products['Item Number'].quantile(0.2)
        ]['Product Name'].tolist()[:3]  # Top 3 expiring soon items

    elif priority == "high_stock_levels":
        # Select items with the highest stock (assuming Item Number correlates with stock)
        shop_priority_items = walmart_products[
            walmart_products['Item Number'] > walmart_products['Item Number'].quantile(0.8)
        ]['Product Name'].tolist()[:3]  # Top 3 high stock items

    elif priority == "promotion_sale":
        # Select items with the highest discount (List Price > Sale Price)
        shop_priority_items = walmart_products[
            walmart_products['List Price'] > walmart_products['Sale Price']
        ]['Product Name'].tolist()[:3]  # Top 3 discounted items

    else:
        shop_priority_items = []  # No priority specified

    return shop_priority_items



def find_best_gap_filling_combination(gap_filling_pool, gap, walmart_products):
    """
    Finds the best combination of items from the gap_filling_pool to reach the gap amount.
    Prioritizes exact matches and then minimizes the difference.
    """
    best_combination = None
    min_difference = float('inf')

    # 1. Check for exact match
    for item in gap_filling_pool:
        product_data = walmart_products[walmart_products['Product Name'] == item]['Sale Price']
        if not product_data.empty:
            item_price = product_data.iloc[0]
            if abs(item_price - gap) < 1e-6:
                return [item]  # Found an exact match!

    # 2. Check small combinations (up to size 3)
    for r in range(1, min(4, len(gap_filling_pool) + 1)):
        for combination in itertools.combinations(gap_filling_pool, r):
            total_price = 0
            valid_combination = True
            for combo_item in combination:
                product_data = walmart_products[walmart_products['Product Name'] == combo_item]['Sale Price']
                if not product_data.empty:
                    item_price = product_data.iloc[0]
                    total_price += item_price
                else:
                    valid_combination = False
                    break

            if not valid_combination:
                continue

            difference = abs(total_price - gap)
            if difference < min_difference:
                min_difference = difference
                best_combination = list(combination)

    return best_combination


import difflib

def find_best_matching_product(product_name, product_list):
    """
    Finds the closest matching product name from the dataset using difflib.
    """
    closest_matches = difflib.get_close_matches(product_name, product_list, n=1, cutoff=0.6)
    return closest_matches[0] if closest_matches else None


def get_smart_cart_recommendations(user_id, current_cart_items, shipping_threshold=20, alpha=0.5,
                                   shop_priority="", max_overspend_percentage=0.2, num_hybrid_recommendations=5):
    """
    Combines collaborative filtering, content-based recommendations, and shop priority items
    while considering the shipping threshold and a maximum overspend percentage.
    """

    # 1. Hybrid Recommendations (CF + Content-Based)
    hybrid_recommendations = get_hybrid_recommendations(
        user_id, current_cart_items,
        num_recommendations=num_hybrid_recommendations,
        alpha=alpha
    )

    # 2. Calculate current cart value
    current_cart_value = 0.0
    # for item in current_cart_items:
    #     # Use fuzzy matching to find the best match in the dataset
    #     matched_product = find_best_matching_product(item, walmart_products['Product Name'].tolist())

    #     if matched_product:
    #         product_data = walmart_products[walmart_products['Product Name'] == matched_product]['Sale Price']
    #     else:
    #         print(f"Warning: Product '{item}' not found in walmart_products. Skipping embedding.")
    #         continue  # Skip this item if no match is found

    #     if not product_data.empty:
    #         current_cart_value += product_data.iloc[0]
    for item in current_cart_items:
      # Try to find the best match for the product
      matched_product = find_best_matching_product(item, walmart_products['Product Name'].tolist())

      if matched_product:
          product_data = walmart_products[walmart_products['Product Name'] == matched_product]['Sale Price']
      else:
          print(f"⚠️ Warning: Product '{item}' not found in walmart_products.")

          # Look for alternative recommendations
          alternative_items = walmart_products[walmart_products['Category'].str.contains("Battery", case=False, na=False)]

          if not alternative_items.empty:
              alternative_product = alternative_items['Product Name'].iloc[0]
              print(f"🔄 Suggesting alternative: {alternative_product}")
              matched_product = alternative_product
              product_data = walmart_products[walmart_products['Product Name'] == matched_product]['Sale Price']
          else:
              print(f"❌ No alternatives found for '{item}'. Skipping.")
              continue  # Skip if no match or alternative is found

    if not product_data.empty:
        current_cart_value += product_data.iloc[0]

    # 3. Calculate gap to reach shipping threshold
    gap = shipping_threshold - current_cart_value

    # 4. Calculate max acceptable cart value
    max_cart_value = shipping_threshold * (1 + max_overspend_percentage)

    # 5. Get shop priority items
    shop_priority_items = get_shop_priority_items(shop_priority)

    # 6. Create Gap-Filling Pool
    gap_filling_pool = walmart_products[walmart_products['Sale Price'] <= gap]['Product Name'].tolist()

    # 7. Final Recommendations List
    final_recommendations = []

    # 8. Attempt to Fill the Gap with Exact or Combinations
    best_gap_fill_items = find_best_gap_filling_combination(gap_filling_pool, gap, walmart_products)

    if best_gap_fill_items:
        final_recommendations.extend(best_gap_fill_items)

    # 9. Add Shop Priority Items (if any and within budget)
    for item in shop_priority_items:
        matched_product = find_best_matching_product(item, walmart_products['Product Name'].tolist())

        if matched_product:
            product_data = walmart_products[walmart_products['Product Name'] == matched_product]['Sale Price']
        else:
            continue  # Skip if no match is found

        if not product_data.empty:
            item_price = product_data.iloc[0]
            total_recommendations = sum(
                walmart_products.loc[walmart_products['Product Name'] == i, 'Sale Price'].iloc[0]
                for i in final_recommendations
                if not walmart_products[walmart_products['Product Name'] == i]['Sale Price'].empty
            )

            # Ensure it does not exceed max cart value
            if (current_cart_value + total_recommendations + item_price) <= max_cart_value:
                final_recommendations.append(matched_product)
            else:
                print(f"Skipping {matched_product} - would exceed max cart value.")

    # 10. Add Hybrid Recommendations if Needed
    remaining_gap = shipping_threshold - (current_cart_value + sum([
        walmart_products.loc[walmart_products['Product Name'] == i, 'Sale Price'].iloc[0]
        for i in final_recommendations if not walmart_products[walmart_products['Product Name'] == i]['Sale Price'].empty
    ]))

    if remaining_gap > 0:
        for rec_item in hybrid_recommendations:
            if remaining_gap > 0:
                matched_product = find_best_matching_product(rec_item, walmart_products['Product Name'].tolist())

                if matched_product:
                    rec_product_data = walmart_products[walmart_products['Product Name'] == matched_product]['Sale Price']
                else:
                    print(f"Skipping {rec_item} - Missing price data.")
                    continue

                # Safe-check for missing values
                if not rec_product_data.empty:
                    rec_item_price = rec_product_data.iloc[0]

                    # Prevent recommending similar items
                    if matched_product not in current_cart_items:
                        if (current_cart_value + rec_item_price) <= max_cart_value:
                            final_recommendations.append(matched_product)
                            remaining_gap -= rec_item_price
                    else:
                        print(f"Skipping {matched_product} - Too similar to an item already in the cart.")

    # Print Final Cart Status
    print("\n=== Smart Cart Recommendation Summary ===")
    print(f"User ID: {user_id}")
    print(f"Current Cart: {current_cart_items}")
    print(f"Cart Amount: {current_cart_value:.2f}")
    print(f"Gap to Free Shipping: {gap:.2f}")
    print(f"Max Cart Value: {max_cart_value:.2f}")
    print(f"Final Recommendations: {final_recommendations}")

    return final_recommendations



# Example Usage
user_id = walmart_sales['UserID'].iloc[0]  # Example user
current_cart = ['Duracell - AAA Batteries (4-Pack)']  # Example cart
shipping_threshold = 20.0  # Example shipping threshold
shop_priority = "promotion_sale"  # Example shop priority
max_overspend_percentage = 0.2  # Example overspend allowed
num_hybrid_recommendations = 5

recommendations = get_smart_cart_recommendations(
    user_id, current_cart, shipping_threshold,
    shop_priority=shop_priority,
    max_overspend_percentage=max_overspend_percentage,
    num_hybrid_recommendations=num_hybrid_recommendations
)
```

```python
import gradio as gr
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
from sklearn.neighbors import NearestNeighbors

# Load datasets (Ensure df and walmart_sales are loaded)
walmart_products = df  # Loaded externally
walmart_sales = walmart_sales  # Loaded externally

# Rename columns
walmart_sales.rename(columns={'User_ID': 'UserID', 'Product_ID': 'ProductID'}, inplace=True, errors='ignore')

# Merge sales and products
sales_data = walmart_sales.merge(walmart_products, left_on='ProductID', right_on='Uniq Id', how='left')

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate product embeddings
product_embeddings = model.encode(walmart_products['combined_text'].tolist())

# Train collaborative filtering model
user_item_matrix = sales_data.pivot_table(index='UserID', columns='ProductID', values='Purchase', fill_value=0)
knn_model = NearestNeighbors(metric='cosine', algorithm='brute')
knn_model.fit(user_item_matrix)

# Recommendation functions
def get_user_recommendations(user_id, num_recommendations=5):
    if int(user_id) not in user_item_matrix.index:
        return ["User ID not found. Enter a valid ID."]
    user_vector = user_item_matrix.loc[int(user_id)].values.reshape(1, -1)
    distances, indices = knn_model.kneighbors(user_vector, n_neighbors=6)
    similar_users = user_item_matrix.iloc[indices[0]].mean()
    similar_users[user_item_matrix.loc[int(user_id)] > 0] = 0
    return similar_users.sort_values(ascending=False).head(num_recommendations).index.tolist()

def get_content_based_recommendations(cart_items, num_recommendations=5):
    if not cart_items:
        return ["Cart is empty. Add items for recommendations."]
    cart_embeddings = model.encode([walmart_products[walmart_products['Product Name'] == item]['combined_text'].iloc[0] for item in cart_items if item in walmart_products['Product Name'].values])
    similarity_scores = cosine_similarity(cart_embeddings, product_embeddings).mean(axis=0)
    recommended_indices = np.argsort(similarity_scores)[::-1]
    return [walmart_products.iloc[idx]['Product Name'] for idx in recommended_indices[:num_recommendations]]

# Gradio UI
def recommend(user_id, cart_items, num_recommendations):
    cf_recommendations = get_user_recommendations(user_id, num_recommendations)
    cb_recommendations = get_content_based_recommendations(cart_items, num_recommendations)
    return cf_recommendations + cb_recommendations

# Create multi-page UI
def login(username, password):
    if username and password:  # Add authentication logic
        return "Welcome! Please proceed to shopping.", "shopping"
    return "Invalid login. Try again.", "login"

def shopping(user_id):
    return f"User {user_id}, select products to add to cart.", "cart"

def cart_summary(user_id, cart_items):
    recommended_products = recommend(user_id, cart_items, 5)
    return f"Cart Summary\nTotal Price: $XXX\nPrice Gap for Free Delivery: $XX\nRecommended Items: {', '.join(recommended_products)}", "checkout"

def checkout():
    return "Purchase successful!", "confirmation"

demo = gr.Blocks()
with demo:
    with gr.Tab("Login"):
        username = gr.Textbox(label="Username")
        password = gr.Textbox(label="Password", type="password")
        login_btn = gr.Button("Login")
        output = gr.Label()
        login_btn.click(login, inputs=[username, password], outputs=[output, gr.State("shopping")])

    with gr.Tab("Shopping"):
        user_id = gr.Textbox(label="User ID")
        cart_items = gr.CheckboxGroup(choices=walmart_products['Product Name'].tolist(), label="Select Products")
        proceed_btn = gr.Button("Proceed to Cart")
        shopping_output = gr.Label()
        proceed_btn.click(shopping, inputs=[user_id], outputs=[shopping_output, gr.State("cart")])

    with gr.Tab("Cart Summary"):
        summary_output = gr.Label()
        checkout_btn = gr.Button("Checkout")
        checkout_btn.click(checkout, inputs=[], outputs=[summary_output, gr.State("confirmation")])

    with gr.Tab("Confirmation"):
        confirmation_msg = gr.Label("Purchase Complete!")

demo.launch(share=True)
```

```python
import gradio as gr
import pandas as pd
import numpy as np

def get_products(category):
    """Fetch 10 products from the selected category."""
    filtered_products = walmart_sales[walmart_sales['Category'] == category]
    return filtered_products['Product_Name'].head(10).tolist()

def update_cart(selected_products):
    """Calculate total price, price gap, and recommendations."""
    total_price = walmart_sales[walmart_sales['Product_Name'].isin(selected_products)]['Purchase'].sum()
    price_gap = max(0, 35 - total_price)  # Free delivery threshold = $35

    # Get recommendations based on rollback and sentiment analysis
    recommendations = walmart_sales[walmart_sales['Category'].isin(selected_products)].sample(5)['Product_Name'].tolist()

    return f"Total: ${total_price:.2f}\nPrice Gap: ${price_gap:.2f}\nRecommended: {', '.join(recommendations)}"

# Gradio UI
def shopping_ui():
    with gr.Blocks() as demo:
        gr.Markdown("## Walmart Shopping Cart")

        with gr.Row():
            category = gr.Radio([
                "Beauty & Personal Care", "Health & Household", "Food & Baby",
                "Sports & Outdoors", "Toys & Home", "Other"
            ], label="Select a Category")

            product_list = gr.CheckboxGroup([], label="Select Products")

            cart_summary = gr.Textbox(label="Cart Summary", interactive=False)

        category.change(lambda cat: get_products(cat), inputs=[category], outputs=[product_list])
        product_list.change(lambda selected: update_cart(selected), inputs=[product_list], outputs=[cart_summary])

    return demo

# Launch UI
demo = shopping_ui()
demo.launch(share=True)
```
